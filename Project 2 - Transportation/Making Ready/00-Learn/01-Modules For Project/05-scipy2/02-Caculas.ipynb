```python
from scipy.misc import derivative  #00000
from scipy.integrate import quad

import matplotlib.pyplot as plt
import numpy as np


def f(x):
    return x ** 2 * np.sin(2 * x) * np.exp(-x)


x = np.linspace(0, 1, 100)
```

```python
plt.plot(x, f(x))
```

```python
a = derivative(f, x, dx=1e-6, n=2)
plt.plot(x, a)
```

```python
from scipy.integrate import dblquad

integrand = lambda x: x ** 2 * np.sin(2 * x) * np.exp(-x)
integral, integral_err = quad(integrand, 0, 1)
```

```python
integral
```

```python
integral_err
```

```python
integrand = lambda x, y: np.sin(x + y ** 2)
lwr_y = lambda x: -x
upr_y = lambda x: x**2
integrate, integrand_err = dblquad(integrand, 0, 1 , lwr_y, upr_y)
integrate
```

```python
from scipy.integrate import odeint
```

```python
def dvdt(v, t):
    return 3*v**2 - 5
v0 = 0
```

```python
t = np.linspace(0,1,100)
sol = odeint(dvdt, v0, t)
```

```python
sol.T[0]
```

```python
plt.plot(t,sol)
```
