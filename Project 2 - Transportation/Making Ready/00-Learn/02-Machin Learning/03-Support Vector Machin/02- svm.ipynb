```python
import pandas as pd
from sklearn.datasets import load_iris; iris = load_iris()
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df["target"] = iris.target
df
```

```python
iris.target_names
```

```python
df[df.target==1]
```

# how to add a column with other column

```python
df["flower_name"] = df.target.apply(lambda x: iris.target_names[x])
df
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
df2
```

# plot some datas

```python
plt.scatter(df0["sepal length (cm)"], df0["sepal width (cm)"], color ="yellow", marker = "o")
plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], color ="green", marker = "o")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
```

```python
plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], color ="green", marker = "o")
plt.scatter(df2["sepal length (cm)"], df2["sepal width (cm)"], color ="blue", marker = "o")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
```

```python
plt.scatter(df2["petal length (cm)"], df2["petal width (cm)"], color ="blue", marker = "o")
plt.scatter(df0["petal length (cm)"], df0["petal width (cm)"], color ="yellow", marker = "o")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
```

# make my svm model

```python
X = df.drop(["target", "flower_name"], axis="columns")
X.head()
```

```python
y = df.target
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)
```

```python
print(len(X_test),len(X_train))
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model = SVC(C=10000, gamma=199, kernel="linear")
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
