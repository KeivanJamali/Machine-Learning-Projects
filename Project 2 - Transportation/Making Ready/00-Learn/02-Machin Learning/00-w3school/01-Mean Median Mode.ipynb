# Mean, Median, and Mode

```python
import numpy as np

speed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]
mean = (99 + 86 + 87 + 88 + 111 + 86 + 103 + 87 + 94 + 78 + 77 + 85 + 86) / 13
mean = np.mean(speed)
print(mean)
speed = [99, 86, 87, 88, 86, 103, 87, 94, 78, 77, 85, 86]
median = np.median(speed)
print(median)
```

```python
from scipy import stats

speed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]
mode = stats.mode(speed)
print(mode)
```

# Standard Deviation

```python
import numpy as np

speed = [86, 87, 88, 86, 87, 85, 86]
stan = np.std(speed)
print(stan)

speed = [32, 111, 138, 28, 59, 77, 97]
stan = np.std(speed)
print(stan)
```

# Percentiles

```python
import numpy as np

ages = [5, 31, 43, 48, 50, 41, 7, 11, 15, 39, 80, 82, 32, 2, 8, 6, 25, 36, 27, 61, 31]
p = np.percentile(ages, 75)
print(p)
ages = [5, 31, 43, 48, 50, 41, 7, 11, 15, 39, 80, 82, 32, 2, 8, 6, 25, 36, 27, 61, 31]
x = np.percentile(ages, 90)
print(x)
```

# Data Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.uniform(0, 5, 250)
# print(data)

plt.hist(data, 15)  # this number is the number of bars
plt.show()
```

```python
x = np.random.uniform(0.0, 5.0, 100000)

plt.hist(x, 100)
plt.show()
```

# Normal Data Distribution

```python
x = np.random.normal(10.0, 1.0, 1000000)  # first number is mean and the second one is std.

plt.hist(x, 100)
plt.show()
```
