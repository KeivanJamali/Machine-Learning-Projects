```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node("Input Layer")
G.add_node("GCN/GAT")
G.add_node("MLP")
G.add_node("Output Layer")

# Add edges
G.add_edge("Input Layer", "GCN/GAT")
G.add_edge("GCN/GAT", "MLP")
G.add_edge("MLP", "Output Layer")

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=12, font_weight="bold")
plt.title("GNN Model Architecture Diagram")
plt.show()
```
