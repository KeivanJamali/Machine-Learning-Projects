```python
import pandas as pd
import numpy as np
import Test as ML
%load_ext autoreload
%autoreload 2
```

```python
# Example of using the FlowDataLoader
base_path = 'data'  # Adjust this to your actual path
city = 'Anaheim'
random = 10
missing_rate = 10  # For example, 10% missing rate

data_loader = ML.FlowDataLoader(base_path, city, random, missing_rate)
x_train, y_train, x_val, y_val, x_test, y_test = data_loader.get_data()
```

```python
# Initialize Random Forest model
rf_model = ML.RandomForestFlowPredictor(n_estimators=100, max_depth=None)

# Train Random Forest model
rf_model.train(x_train, y_train)

# Evaluate on validation data
rmse_val, mae_val, r2_val = rf_model.evaluate(x_val, y_val)
print(f"Validation - RMSE: {rmse_val}, MAE: {mae_val}, R2: {r2_val}")

# Evaluate on test data
rmse_test, mae_test, r2_test = rf_model.evaluate(x_test, y_test)
print(f"Test - RMSE: {rmse_test}, MAE: {mae_test}, R2: {r2_test}")
```

```python
# Loop through cities and missing rates to get results
base_path = 'data'  # Adjust this to your actual path
cities = ['Anaheim', 'SiouxFalls', 'Chicago_Sketch']
randoms = [0, 5, 10, 15, 20, 25, 30, 35, 40, 50]
missing_rates = [10, 20, 30, 40, 50, 60, 70, 80, 90, 95, 97]

results = []

for city in cities:
    for random in randoms:
        for missing_rate in missing_rates:
            # Load data
            data_loader = ML.FlowDataLoader(base_path, city, random, missing_rate)
            x_train, y_train, x_val, y_val, x_test, y_test = data_loader.get_data()
            
            # Initialize and train the model
            predictor = ML.RandomForestFlowPredictor()
            predictor.train(x_train, y_train)

            # Evaluate the model on test data
            rmse, mae, r2 = predictor.evaluate(x_test, y_test)

            # Append results
            results.append({
                'City': city,
                'random': random,
                'Missing Rate': missing_rate,
                'RMSE': rmse,
                'MAE': mae,
                'R2': r2
            })

# Convert results to a DataFrame for better visualization
results_df = pd.DataFrame(results)
print(results_df)
```

```python
results_df.to_csv("result.csv")
```
