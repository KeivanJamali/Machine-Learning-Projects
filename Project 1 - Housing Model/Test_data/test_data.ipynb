```python
import pandas as pd
from Statistical_Hypothesis_Tests import Correlation

data = pd.read_csv("Housing.csv")
data.replace("yes", 1,inplace=True)
data.replace("no", 0, inplace=True)
data.replace("furnished", 1, inplace=True)
data.replace("semi-furnished", 0.5, inplace=True)
data.replace("unfurnished", 0, inplace=True)
corr = Correlation()


for i in range(1, len(data.columns)):
    print(f"Prices VS {data.columns[i]}")
    print("monotonic")
    corr.spearman(data.price, data.iloc[:,i])
    print("linear")
    corr.pearson(data.price, data.iloc[:,i])
    print()
    print()
    print()
```
