```python
mean = [0, 0]
cov = [[1, -0.5], [-0.5, 2]]
y = np.random.multivariate_normal(mean, cov, 1000)
plt.scatter(y[:, 0], y[:, 1], marker=".")
y
```

```python
size = len(y)
mean = [0, 0]
for i, j in y:
    mean[0] += i
    mean[1] += j
mean[0] /= size
mean[1] /= size

cov = 0
for i, j in y:
    cov += (i-mean[0]) * (j-mean[1])
cov /= size

mean2 = np.mean(y, axis=0)
cov2 = np.cov(y, rowvar=False)
```

```python
y_centered = y-mean
y_centered.T.dot(y_centered)/(size-1)
```
