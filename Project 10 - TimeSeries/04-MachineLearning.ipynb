## Airline - No diffrencing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_absolute_percentage_error, r2_score
```

```python
data = pd.read_csv("raw_data/airline_passengers.csv", index_col="Month", parse_dates=True)
```

```python
data["LogPassengers"] = np.log(data["Passengers"])
```

```python
Ntest = 12
train = data.iloc[:-Ntest]
test = data.iloc[-Ntest:]
```

```python
series = data["LogPassengers"].to_numpy()

T = 10
X = []
Y = []
for t in range(len(series) - T):
    x = series[t:t+T]
    X.append(x)
    y = series[t+T]
    Y.append(y)
```

```python
X = np.array(X).reshape(-1, T)
Y = np.array(Y)
N = len(X)
print(f"X.shpae: {X.shape}, Y.shape: {Y.shape}")
```

```python
x_train, y_train = X[:-Ntest], Y[:-Ntest]
x_test, y_test = X[-Ntest:], Y[-Ntest:]
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
train_idx = data.index <= train.index[-1]
test_idx = ~train_idx
train_idx[:T] = False
```

```python
data.loc[train_idx, "LR_1step_train"] = lr.predict(x_train)
data.loc[test_idx, "LR_1step_test"] = lr.predict(x_test)
data[["LogPassengers", "LR_1step_train", "LR_1step_test"]].plot(figsize=(15, 5));
```

```python
multistep_predictions = []

last_x = x_test[0]
while len(multistep_predictions) < Ntest:
    pred = lr.predict(last_x.reshape(1, -1))[0]
    multistep_predictions.append(pred)
    last_x = np.roll(last_x, -1)
    last_x[-1] = pred
```

```python
data.loc[test_idx, "LR_multistep"] = multistep_predictions
data[["LogPassengers", "LR_1step_train", "LR_1step_test", "LR_multistep"]].plot(figsize=(15, 5));
```

```python
series = data["LogPassengers"].to_numpy()

Tx = T
Ty = Ntest
X = []
Y = []
for t in range(len(series) - Tx - Ty + 1):
    x = series[t:t+Tx]
    X.append(x)
    y = series[t+Tx:t+Tx+Ty]
    Y.append(y)

X = np.array(X).reshape(-1, Tx)
Y = np.array(Y).reshape(-1, Ty)
N = len(X)
print(f"X.shape: {X.shape}, Y.shape: {Y.shape}")
```

```python
x_train_m, y_train_m = X[:-1], Y[:-1]
x_test_m, y_test_m = X[-1:], Y[-1:]
```

```python
lr = LinearRegression()
lr.fit(x_train_m, y_train_m)
lr.score(x_train_m, y_train_m)
```

```python
lr.score(x_test_m, y_test_m)
```

```python
r2_score(lr.predict(x_test_m).flatten(), y_test_m.flatten())
```

```python
data.loc[test_idx, "LR_multioutput"] = lr.predict(x_test_m).flatten()
```

```python
cols = ["LogPassengers", "LR_1step_train", "LR_1step_test", "LR_multistep", "LR_multioutput"]
data[cols].plot(figsize=(15, 5));
```

```python
mape1 = mean_absolute_percentage_error(y_test, multistep_predictions)
print(mape1)

mape2 = mean_absolute_percentage_error(y_test, data.loc[test_idx, "LR_multioutput"])
print(mape2)
```

```python
def one_step_and_multistep_forecast(model, name):
    model.fit(x_train, y_train)
    print(f"One-step foecast: {name}")
    # print(f"Train r2: {model.score(x_train, y_train):.3f}")
    # print(f"Test r2: {model.score(x_test, y_test):.3f}")

    data.loc[train_idx, f"{name}_1step_train"] = model.predict(x_train)
    data.loc[test_idx, f"{name}_1step_test"] = model.predict(x_test)

    multistep_predictions = []

    last_x = x_test[0]
    while len(multistep_predictions) < Ntest:
        pred = model.predict(last_x.reshape(1, -1))[0]
        multistep_predictions.append(pred)
        last_x = np.roll(last_x, -1)
        last_x[-1] = pred
    data.loc[test_idx, f"{name}_multistep_test"] = multistep_predictions

    mape = mean_absolute_percentage_error(y_test, multistep_predictions)
    print(f"MAPE: {mape:.3f}")

    cols = ["LogPassengers", f"{name}_1step_train", f"{name}_1step_test", f"{name}_multistep_test"]
    data[cols].plot(figsize=(15, 5));
```

```python
def one_step_and_multistep_forecast(model, name):
    model.fit(Xtrain, Ytrain)
    print(f"One-step forecast: {name}")
    # print(f"Train R^2: {model.score(Xtrain, Ytrain)}")
    # print(f"Test R^2 (1-step): {model.score(Xtest, Ytest)}")

    # store 1-step forecast

    data.loc[train_idx, f"{name}_1step_train"] = prev[train_idx] + model.predict(Xtrain)
    data.loc[test_idx, f"{name}_1step_test"] = prev[test_idx] + model.predict(Xtest)

    # generate multi-step forecast
    multistep_predictions = []

    # first test input
    last_x = Xtest[0]

    while len(multistep_predictions) < Ntest:
        p = model.predict(last_x.reshape(1, -1))[0]

        # update the predictions list
        multistep_predictions.append(p)

        # make the new input
        last_x = np.roll(last_x, -1)
        last_x[-1] = p

    # stoe multi-step forecast
    data.loc[test_idx, f"{name}_multistep_test"] = last_train + np.cumsum(multistep_predictions)

    # MAPE of multi-step forecast
    mape = mean_absolute_percentage_error(test_log_pass, data.loc[test_idx, f"{name}_multistep_test"])
    print("test MAPE (multi-step):", mape)

    # plot 1-step and multi-step forecast
    cols = ["LogPassengers", f"{name}_1step_train", f"{name}_1step_test", f"{name}_multistep_test"]

    data[cols].plot(figsize=(15, 5))
```

```python
one_step_and_multistep_forecast(SVR(), "SVR")
```

```python
one_step_and_multistep_forecast(RandomForestRegressor(), "RF")
```

```python
def multi_output_forecast(model, name):
    model.fit(Xtrain_m, Ytrain_m)

    # save multi-output forecast to dataframe
    data.loc[test_idx, f"{name}_multioutput"] = last_train + np.cumsum(model.predict(Xtest_m).flatten())
    mape = mean_absolute_percentage_error(test_log_pass, data.loc[test_idx, f"{name}_multioutput"])
    print(f"Test Mape (multi-output): {mape}")

    # plot all forecast
    # assume that previous functions have been run

    cols = ["LogPassengers",
            f"{name}_1step_train",
            f"{name}_1step_test",
            f"{name}_multistep_test",
            f"{name}_multioutput"]
    data[cols].plot(figsize=(15, 5));
```

```python
class SVRWrapper:
    def __init__(self, h, **args):
        self.h = h
        self.models = [SVR(**args) for _ in range(h)]

    def fit(self, X, Y):
        for k in range(self.h):
            self.models[k].fit(X, Y[:, k])

    def predict(self, X):
        p = [m.predict(X) for m in self.models]
        return np.hstack(p)
```
